```python
from timbre_recognition.utils.io import *
import numpy as np
import matplotlib.pyplot as plt
import librosa
import librosa.display
import time
filename = './trumpet-g4.wav'
start_time = time.time()
data, fs = read_audio_wav_tf(filename, desired_channels=1)
```

```python
data = data[:, 0]
t = np.arange(len(data))
fig, (ax1, ax2) = plt.subplots(nrows=2)
ax1.plot(t, data)
ax2.specgram(data, NFFT=1024, Fs=fs)
plt.show()
```

```python
y, sr = librosa.load('timbre_recognition/datasets/ESC-50/audio/1-977-A-39.wav', sr=None, mono=True)
librosa.output.write_wav('977-44.wav', y, sr)
S = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=128,
                                   fmax=8000)
plt.figure(figsize=(10, 4))
librosa.display.specshow(librosa.power_to_db(S,
                         ref=np.max), y_axis='mel', 
                         fmax=8000, x_axis='time',
                         cmap='viridis') # default is magma
plt.colorbar(format='%+2.0f dB')
# plt.axis('off') # uncomment to remove axes
plt.title('Mel spectrogram')
plt.tight_layout()
```
